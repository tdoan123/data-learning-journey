```python
url = "https://raw.githubusercontent.com/JosephBARBIERDARNAL/data-matplotlib-journey/refs/heads/main/us-counties/us-counties.geojson"
```

```python
url_country = "https://raw.githubusercontent.com/JosephBARBIERDARNAL/data-matplotlib-journey/refs/heads/main/world/world.geojson"
```

```python
import geopandas as gdp
import matplotlib.pyplot as plt 
```

```python
world = gdp.read_file(url_country)
```

```python
world.plot()
```

```python
world
```

```python
len(world['name'].unique())
```

```python
len(world[world['continent']=='Europe'])
```

```python
world.plot()
```

```python
fig, ax = plt.subplots()
world.plot(ax=ax)
plt.show() 
```

```python
# option: tweak limits of ax 
fig, ax = plt.subplots()
world.plot(
    column='name',
    ax = ax
)
ax.set_xlim(-10,20)
ax.set_ylim(30,60)

plt.show()
```

```python
# select row from the data  
vietnam = world[world['name'] == 'Vietnam']

fig, ax = plt.subplots() 

vietnam.plot(
    column = 'name',
    ax = ax 
)

plt.show()
```

```python
# option: tweak limits of ax; then customize the chart   
fig, ax = plt.subplots()
world.plot(
    column='name',
    ax = ax,
    color = 'lightgray',
    edgecolor = 'black'
)
ax.set_xlim(-10,20)
ax.set_ylim(30,60)

plt.show()
```

```python
europe = world[world['continent'] == 'Europe']
```

```python
asia = world[world['continent'] == 'Asia']
```

```python
fig, axs = plt.subplots(ncols = 2)
europe.plot(ax=axs[0], color = 'blue', edgecolor = 'black')
asia.plot(ax=axs[1], color = 'green', edgecolor = 'black')
axs[0].text(x=-50, y=20, s='Europe')
axs[1].text(x=20, y=0, s='Asia')
plt.show()
```

```python
# to highlight a specific country   

vietnam = world[world['name'] == 'Vietnam'] # define country   
asia = world[world['continent'] == 'Asia'] # define the region  

fig, ax = plt.subplots() 
asia.plot(color='lightgray', edgecolor='grey', ax=ax)
vietnam.plot(color='red', edgecolor='red', ax=ax)
ax.set_xlim(80,130)
ax.set_ylim(-10, 40)
ax.axis('off')
plt.show()
```

```python
fig, ax = plt.subplots()
world.plot(ax = ax)
ax.axis('off')
plt.show() 
```

```python
# zoom in on South America 
south_america = world[world['name'].isin(('Brazil', 'Argentina', 'Peru', 'Colombia', 'Bolivia', 'Paraguay'))]
```

```python
south_america
```

```python
fig, ax = plt.subplots()
fig.set_facecolor('lightgrey')
ax.set_facecolor("lightgrey")
ax.axis('off')

south_america.plot(
    color= 'darkred',
    edgecolor = 'black',
    linewidth = 1.2,
    ax = ax  
)


plt.show()
```

```python
japan = world[world['name'] == 'Japan']
germany = world[world['name'] == 'Germany']
peru = world[world['name'] == 'Peru']
australia = world[world['name'] == 'Australia']
```

```python
fig, axs = plt.subplots(nrows= 2, ncols=2)
japan.plot(
    ax = axs[0, 0], 
    color = 'red'
    )
germany.plot(
    ax = axs[0, 1], 
    color = 'purple'
    )
peru.plot(
    ax = axs[1, 0], 
    color = 'green'
    )
australia.plot(
    ax = axs[1, 1], 
    color = 'darkblue'
    )

plt.show()
```
