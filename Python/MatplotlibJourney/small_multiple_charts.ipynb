```python

fig, ax = plt.subplots()
ax.plot(
    df['date'],
    df[df['country'] == 'united states'],
)

plt.show()
```

```python
# setting layout  
fig, axs = plt.subplots(
    nrows = 3,
    ncols = 3,
    layout = 'tight'
)

countries = df['country'].unique()

for focus_country, ax in zip(countries, axs.flat):
    for i,country in enumerate(countries):
        subset = df[df.country == country]
        if focus_country == country:
            zorder = 10
            color = colors[i]
            lw = 2 
            ax.text(x = '2024-01-01', y = 110,
                    s=focus_country.title(),
                    size = 10, ha = 'right',
                    color = color,
                    fontweight = 'bold')
        else:
            color = 'darkgrey'
            lw = 0.3
            zorder = -10
        ax.plot(subset["date"], subset["consumer confidence"], color = color, lw = lw, zorder = zorder)
    
    ax.grid(axis = "y", lw = 0.3)
    ax.spines[["top", "right", "bottom"]].set_visible(False)
    ax.tick_params(length = 0, labelsize= 6)
    ax.set_xticks(['2020-01-01', '2022-01-01', '2024-01-01'],
                labels = [2020,2022,2024])
    ax.set_ylim([-60,130])
    ax.set_yticks([-50,0,50,100])
    ax.set_facecolor(background_color)
fig.set_facecolor(background_color)
plt.show()
```
