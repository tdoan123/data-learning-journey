## Figure and Axes 

`Axes` is where the actual data visualization happens. It contains the plot elements such as lines, bars, scatter points, and more  

`Figure`: is a top-level container for everhything in a Matplotlib chart. Thnk of it as the canvas where all elements are drawn, including one or more Axes. 
Also. `Figure`  holds other components such as titles, legends, and global labels that are not tied to a specific Axes 

## One figure can have several axes  

## Use case: adding text  

`text()` function: require at least three arugments: 

`x`: the position on the x-axis

`y`: the position on the y-axis 

`s`: the text content 

this function can be used in two different ways: 

ax.text(x, y, s): adds text to the Axes 

fig.text(x, y, s): adds text to the Figure 

## Axes and Figure: Different Coordinate Systems

- Axes use the same coordinate system as the data (i.e. data coordinates)
- Figure coordinates range from 0 to 1 on both the x and y axes    

### Axes

```python
fig, ax = plt.subplots()
ax.plot([1,2,3,4,5], [2,3,1,4,5])
ax.text(3, 3, "Hello world!")

plt.show()
```

### Figure

```python
fig, ax = plt.subplots()

ax.plot([1,2,3,4,5], [2,3,1,4,5])
fig.text(0.5, 0.5, "Hello world!")

plt.show()
```

### **Key take away**
1)  Every Matplotlib visualization is a Figure with one or more Axes.
2) A Figure is the overall container for all elements, including Axes and other graph components. Its coordinates range from 0 to 1 on both x and y axes.
3) An Axes is a container for displaying a single chart (line chart, bar plot, etc.), with coordinates in the same scale as the data.
4) When a Figure has a single Axes (as in every chart we've made so far), the center of the Axes and the Figure align.

### Plotting

the plotting group is the most important. It includes fundamental functions for creating chart types, like hist() for histograms, scatter() for scatterplots,bxp() for boxplots, and more.

### Text and Annotations

the second most important group. Annotation is critical in data visualization, and we'll spend a lot of time adding text (e.g., ax.text), arrows, lines, and other shapes to enhance the insights a graph provides.

use any font of our choice with pyfonts
create arrows in virtually any style with drawarrow

### Axis limits, direction and scale

the third most important group: everything related to axes scale and limits.

These functions not only manage the x and y axis limits but also allow us to change the scale to logarithmic, quasi-logarithmic, logit, and more.


```python
# import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 5, 3, 4, 2]

fig, ax = plt.subplots(figsize=(4,3))
ax.plot(
  x, y,
  color="lightgrey",
  linewidth=2,
  zorder=1
)
ax.scatter(
  x, y,
  color="#69b3a2",
  linewidth=1,
  edgecolor="black",
  zorder=2
)
plt.show()

# Combining two chart type, I have the line chart with dots to show the figures 
```

```python
# Barplot 

y = ["France", "Spain", "Canada", "UK", "Australia"]
values = [2, 5, 3, 4, 5]

fig, ax = plt.subplots()
ax.barh(y, values)
ax.set_title('Barplot Chart')
plt.show()
```

```python
# Area chart

x = [1, 2, 3, 4, 5, 6, 7]
y = [2, 5, 3, 4, 5, 2, 6]

fig, ax = plt.subplots()
ax.fill_between(x, y)
ax.set_title('Area Chart')
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6, 7]
y = [2, 5, 3, 4, 5, 2, 6]
s = [100, 200, 300, 400, 200, 100, 500]

fig, ax = plt.subplots()
ax.scatter(x, y, s=s)
ax.set_title('Bubble Chart')
plt.show()
```

```python
# Lollipop plot 
y = [2, 5, 3, 4, 5, 2, 6]

fig, ax = plt.subplots()
ax.stem(y)
ax.set_title('Lollipop Chart')
plt.show()
```

```python
# Pie chart

x = [6, 2, 3, 4, 5]
labels = ["matplotlib", "ggplot2", "plotnine", "D3.js", "PowerBI"]

fig, ax = plt.subplots()
ax.pie(x, labels=labels)
ax.set_title('Pie Chart')
plt.show()
```
