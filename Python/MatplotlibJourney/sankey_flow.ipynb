```python
import pandas as pd
from pySankey.sankey import sankey
```

```python
url = "https://raw.githubusercontent.com/anazalea/pySankey/master/pysankey/fruits.txt"

df = pd.read_csv(url, sep = ' ', names=['true','predicted'])
```

```python
df
```

```python

colorDict = {
    'apple':'#f71b1b',
    'blueberry':'#1b7ef7',
    'banana':'#f3f71b',
    'lime':'#12e23f',
    'orange':'#f78c1b'
}
```

```python
sankey(left= df['true'],right=df['predicted'],aspect=5000, colorDict = colorDict, fontsize = 12, figure_name='fruit')
```

```python
sankey(left= df['true'],right=df['predicted'],aspect=10, colorDict = colorDict, fontsize = 12, figure_name='fruit')
```
