First, we create the list of tuples. Each tuple’s first element contains the text up to the final . in the column name, and the second element is the word following that final .. We can do this using Python’s str.rsplit method, which works similarly to str.split but from the right rather than the left. By itself, str.rsplit won’t make a difference. But if we pass a second, integer argument of 1, it returns a list of two elements split from the final .

````python
general_columns = ['age',
                   'are.you.datascientist',
                   'is.python.main',
                   'company.size',
                   'country.live',
                   'employment.status',
                   'first.learn.about.main.ide',
                   'how.often.use.main.ide',
                   'is.python.main',
                   'main.purposes'
                   'missing.features.main.ide'
                   'nps.main.ide',
                   'python.version.most',
                   'python.years',
                   'python2.version.most',
                   'python3.version.most',
                   'several.projects',
                   'team.size',
                   'use.python.most',
                   'years.of.coding'
                  ]
````

````python
def column_multi_name(column_name):
    if column_name in general_columns:
        return ('general', column_name)
    else:
        first, rest = column_name.rsplit('.', 1)
        return (first, rest)
````

````python
(
    pd
    .MultiIndex.from_tuples([
        column_multi_name(one_column_name)
        for one_column_name in py_df.columns])
)
````

````python
py_df.columns = (
    pd
    .MultiIndex.from_tuples([
    column_multi_name(one_column_name)
    for one_column_name in py_df.columns])
)
````

# Which country has the greatest proportion of Python developers with 11+ years of experience?

````python
# this question is to ask for the proportion of python developers with 11+ years of experience. The above answer was only address the number of developers with 11+ years of experience by each countries.
# to solve this. I need to have the total number of developers represented each countries (this info will have different level of year of python experience) then divided the number of developers with 11+ years of experience to the total number of developers of each country
````

````python
country_experience = (
    py_df['general']
    [['country.live', 'python.years']]
)

all_per_country = (
    country_experience
    ['country.live']
    .value_counts()
)
````

````python
expert_per_country = (
    country_experience
    .loc[country_experience['python.years'] == '11+ years',
         'country.live']
    .value_counts()
)
````

````python
expert_per_country
````

````python
(expert_per_country / all_per_country
    ).sort_values(ascending=False).dropna().head(10)
````

# In YearsCode, replace “Less than 1 year” with 0 and “More than 50 years” with 51

````python
so_df.loc[so_df['YearsCode'] == 'Less than 1 year', 'YearsCode'] = 0 
so_df.loc[so_df['YearsCode'] == 'More than 50 years', 'YearsCode'] = 51
````

````python
so_df['YearsCode'].sort_values()
````

````python
so_df['YearsCode'] = so_df['YearsCode'].astype(int)
````

````python
so_df['YearsCode'].value_counts().sort_values(ascending=False)
````

````python
so_df['experience'] = pd.cut(so_df['YearsCode'],
       bins=[-1, 1, 2, 5, 10, 100],
      labels=['Less than 1 year',
      '1-2 years',
      '3-5 years',
      '6-10 years',
      '11+ years'])
````
