````python
df = (
    df
    .loc[df['CompTotal'] < 2_000_000]
)
````

````python
(
    df['LanguageHaveWorkedWith']
    .str.get_dummies(sep=';')
)
````

````python
df = (
    pd.concat([df,
               df['LanguageHaveWorkedWith']
               .str.get_dummies(sep=';')],
               axis='columns')
)
````

2) I can compare the values of 2 different index assumming they have the same lenghtby using method `.intersection()`  or `.isin()` or `~.isin()` for values that are and are not in the comparing index 

3) After using `str.split()` and `.explode()` to separate the values, I then can combine the the values back and check for the mode (most occurance) of the value by join Index values with the df column. 

