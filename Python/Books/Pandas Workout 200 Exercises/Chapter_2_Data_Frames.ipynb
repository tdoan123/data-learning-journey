### Concept that i need to aware of: 
`DataFrame` return a new dataframe based on two dimensional data  - `DataFrame([[10, 20], [30, 40], [50, 60]])`

`s.loc` accesses elements of a series by labels or a boolean array - `s.loc['a']`

`df.loc` accesses one or more rows of a data frame via the index - `df.loc[5]`

`s.iloc` accesses elements of a series by position - `s.iloc[0]`

`df.iloc` accesses one ore more rows of a data frame by position - `df.iloc[5]`

`[]` access one or more columns in a data frame `df['a']`

`df.assign` add one or more columns to a dataframe `df.assign(a=df['x']*3)`

`str.format` method that works much like f-string `ab{0}'.format(5)

`s.quantile` gets the value at a particular percentage of the values `s.quantile(0.25)`

`pd.concat` join together two data frames `df= pd.concat([df,new_products])`

`df.query` write an sql like query `df.query('v > 300')`

`pd.read_csv` return a new series based on a single column file `s = pd.read_csv('filename.csv').squeeze()`

`interpolate` return a new data frame with NaN values interpolated `df = df.interpolate()` 

`df.dropna` return a new data frame without any 'NaN' values `df.dropna()` 

`s.isin` return a boolean series indicating whether each element of a series is in the provided argument `s.isin([10,20,30])`
