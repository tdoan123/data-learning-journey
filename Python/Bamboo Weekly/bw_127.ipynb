```python
# step 2: add the average line information 
x = work_df2_2024[2024]
y = work_df2_2024['country_name']
facec = 'bisque'
text_color = 'black'

fig, ax = plt.subplots(layout='tight')

fig.set_facecolor(facec)
ax.set_facecolor(facec)

ax.stem(y,x, orientation='horizontal', basefmt=facec, markerfmt='green')

ax.axline((mean2024,0),(mean2024,2),ls='--', color = text_color)

plt.show()
```

```python
# step 3: add the name 
x = work_df2_2024[2024]
y = work_df2_2024['country_name']
facec = 'bisque'
text_color = 'black'

fig, ax = plt.subplots(figsize=(10, 13))

fig.set_facecolor(facec)
ax.set_facecolor(facec)

ax.stem(y,x, orientation='horizontal', basefmt=facec, markerfmt='blue')

ax.axline((mean2024,0),(mean2024,2),ls='--', color = text_color)

# formatting chart 
ax.set_xticks(ticks=[]) # removing x line
ax.set_yticks(ticks=[])  # removing y line 
ax.spines[:].set_visible(False) # removing  figure lines (top, right, bottem, left)

# setting the chart text:  
    # chart title:  
ax.text(
    x = 0, 
    y = 32,
    s='Europe Countries Average Working Hour in 2024',
    size = 16, 
    fontweight = 'bold'
)

    # global average mean line  
ax.text(
    x = mean2024 + 1, 
    y =  10, 
    s = f"Overrall avg. ${mean2024:.1f}",  
    color = text_color, 
    fontsize = 12, 
    rotation = 270,
    ha = 'center'
)

    # to create name and value for each line of the main data  
for i, row in work_df2_2024.iterrows():
    ax.text(
        x = row[2024] - 6,  
        y =  i + 0.1, 
        s = f"{row[2024]:.1f}",
        color = text_color, ha = 'left',
        size = 12
    )
    ax.text(
        x = 0, 
        y = i + 0.1,  
        s = row['country_name'],
        color = text_color,
        size = 12
    )

plt.tight_layout()
plt.show()
```
