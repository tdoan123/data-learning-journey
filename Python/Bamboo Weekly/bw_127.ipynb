## In 2024 (the most recent year for which we have data), which five countries worked the greatest and least number of hours? Which five countries have most increased the number of hours worked (as a percentage) since 2015, and which have decreased?

```python
(
    work_df[2024]
)
```

```python
(
    work_df
    [2024]
    .agg(['nlargest','nsmallest'])
)
```

```python
(
    work_df
    [[2015, 2024]]
    .pct_change(axis='columns', fill_method=None)
    [2024]
    .agg(['nlargest', 'nsmallest'])
)
```

```python
(
    work_df
    [[2015, 2024]]
    .pct_change(axis='columns', fill_method=None)
    [2024]
    .agg(['nlargest', 'nsmallest'])
    .map(lambda x: f'{x:0.2%}')
)
```

## Calculate the percentage change in hours worked for each European country, in each three-year period. That is, calculate the mean number of hours worked in each country for each three-year period

```python
(
    work_df
    .iloc[2:]
)
```

```python
(
    work_df
    .iloc[2:]
    .set_axis(pd.to_datetime(work_df.columns, format='%Y'), axis='columns')    
)
```

```python
(
    work_df
    .iloc[2:]
    .set_axis(pd.to_datetime(work_df.columns, format='%Y'), axis='columns')    
    .T
    .resample('3YE').mean()
    .
)
```

```python
(
    work_df
    .iloc[2:]
    .set_axis(pd.to_datetime(work_df.columns, format='%Y'), axis='columns')    
    .T
    .resample('3YE').mean()
    .T
    .pct_change(axis='columns', fill_method=None)
    .map(lambda x: f'{x:0.2%}')
)
```
