## Calculate the percentage change in hours worked for each European country, in each three-year period. That is, calculate the mean number of hours worked in each country for each three-year period

```python
(
    work_df
    .iloc[2:]
)
```

```python
(
    work_df
    .iloc[2:]
    .set_axis(pd.to_datetime(work_df.columns, format='%Y'), axis='columns')    
)
```

```python
(
    work_df
    .iloc[2:]
    .set_axis(pd.to_datetime(work_df.columns, format='%Y'), axis='columns')    
    .T
    .resample('3YE').mean()
)
```

```python
(
    work_df
    .iloc[2:]
    .set_axis(pd.to_datetime(work_df.columns, format='%Y'), axis='columns')    
    .T
    .resample('3YE').mean()
    .T
    .pct_change(axis='columns', fill_method=None)
    .map(lambda x: f'{x:0.2%}')
)
```
