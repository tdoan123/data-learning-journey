```python
country_df.info()
```

```python
url_country = "https://raw.githubusercontent.com/JosephBARBIERDARNAL/data-matplotlib-journey/refs/heads/main/world/world.geojson"
```

```python
world = gdp.read_file(url_country)
```

```python
europe = world[world['continent'] == 'Europe']
```

```python
df_merge = europe.merge(
    country_df,
    left_on = 'name',
    right_on = 'country'
)
```

```python
df_merge
```

```python
df_merge.sort_values(by="2024_avg_wk_hour" )
```

```python
fig, ax = plt.subplots()
cmap = load_cmap('Bay', cmap_type='continuous')
df_merge.plot(
    ax = ax,
    edgecolor='lightgrey',  
    linewidth=.5,
    column = '2024_avg_wk_hour',
    legend=True,
    legend_kwds={'shrink':0.5},
    cmap= cmap
)
ax.text(x= -10, y = 56, s = 'Netherlands' )
ax_arrow(
    head_position =[5,52],
    tail_position = [-2,56],
    color = 'black',
    radius = 0.1, 
    fill_head= False
)

ax.text(x= 25, y = 38, s = 'Serbia' )
ax_arrow(
    head_position =[22,43],
    tail_position = [30,40],
    color = 'black',
    radius = 0.1, 
    fill_head= False
)

ax.text(x=-10, y= 83, s='2024 Average Weekly Working Hours in Europe')
ax.axis('off')
ax.set_xlim(-10)
ax.set_ylim(30)

plt.show()
```
