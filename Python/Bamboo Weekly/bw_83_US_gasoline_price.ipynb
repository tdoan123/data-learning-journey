# Q2: In which region of the US have gas prices been, on average, the highest since 2000?

## approach:  
- get data from 2000. check for index type to insure the index is DateTime  
- get the mean, and high since 2000  

```python
(
    df
    .loc['2000':]
    .mean()
)
#=> to get all the mean of each column within the index range from 2000 and onward   
```

```python
df_2000 = df.loc['2000':]
```

```python
df_2000 
```

```python
df_2000.mean().nlargest(10)
```

```python
df_2000.mean().sort_values(ascending=False).head(10)
```
