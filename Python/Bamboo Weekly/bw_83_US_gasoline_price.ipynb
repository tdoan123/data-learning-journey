## approach:

```python
df_2000 = df.loc['2000':]
```

```python
df_2000 
```

```python
df_2000.mean().nlargest(10)
```

```python
df_2000.mean().sort_values(ascending=False).head(10)
```

```python
df_2000_mean = df_2000.mean()
```

```python
df_2000_mean 
```

```python
df_2000_mean_gas_price = df_2000_mean.to_frame(name='avg_gas_price')
```

```python
df_2000_mean_gas_price.reset_index(inplace=True)
```

```python
df_2000_mean_gas_price = df_2000_mean_gas_price.sort_values('avg_gas_price',ascending = False)
```

```python
df_2000_mean_gas_price.reset_index(drop=True, inplace=True)
```

```python
mean_2000 = df_2000_mean_gas_price['avg_gas_price'].mean()
```

```python
df_2000_mean_gas_price
```

```python
df_2000_mean_gas_price_by_dd = df_2000_mean_gas_price[df_2000_mean_gas_price['index'].isin(['West Coast', 'Rocky Mountain', 'Gulf Coast', 'Midwest', 'East Coast'])]	
df_2000_mean_gas_price_by_dd
```

```python
df_2000_mean_gas_price_by_dd_mean = df_2000_mean_gas_price_by_dd['avg_gas_price'].mean()
df_2000_mean_gas_price_by_dd_mean
```

```python
df_2000_mean_gas_price_by_dd.reset_index(drop=True, inplace=True)
```

```python
y = df_2000_mean_gas_price_by_dd['index']
x = df_2000_mean_gas_price_by_dd['avg_gas_price']

facec = 'bisque'
text_color = 'black'


fig, ax  = plt.subplots()

fig.set_facecolor(facec)
ax.set_facecolor(facec)

ax.stem(y, x, orientation ='horizontal', basefmt = facec, markerfmt='blue')
ax.axline((df_2000_mean_gas_price_by_dd_mean,0),(df_2000_mean_gas_price_by_dd_mean,3), ls='--', color = text_color)

# chart formating  
ax.set_xticks(ticks=[])
ax.set_yticks(ticks=[])
ax.spines[:].set_visible(False)


# to create name and value for each line of the main data 
for i, row in df_2000_mean_gas_price_by_dd.iterrows(): 
    ax.text(
        x = row['avg_gas_price'] - .2, 
        y = i+.2,
        s = f"{row['avg_gas_price']:.2f}",
        color = text_color,
        ha = 'left',
        size = 12 
    )

    ax.text(
        x = 0, 
        y = i+.2,  
        s = row['index'], 
        color = text_color, 
        size = 12 
    )

# to create texts providing information 
    # chart title:  
ax.text(
    x = 0, 
    y = 5.5,
    s=' ',
    size = 16, 
    fontweight = 'bold'
)
ax.text(
    x = 0, 
    y = 5,
    s='Regional Gasoline Prices: \nCost per Gallon by Defense District',
    size = 14, 
    fontweight = 'bold'
)

# average mean line    
ax.text(
    x = df_2000_mean_gas_price_by_dd_mean + .2, 
    y =  1, 
    s = f"Overrall avg. ${df_2000_mean_gas_price_by_dd_mean:.2f} /gallon",  
    color = text_color, 
    fontsize = 12, 
    rotation = 270,
    ha = 'center'
)
plt.show()
```

```python
y = df_2000_mean_gas_price['index']
x = df_2000_mean_gas_price['avg_gas_price']

facec = 'bisque'
text_color = 'black'


fig, ax  = plt.subplots(layout = 'tight')

fig.set_facecolor(facec)
ax.set_facecolor(facec)

ax.stem(y, x, orientation ='horizontal', basefmt = facec, markerfmt='blue')
ax.axline((mean_2000,0),(mean_2000,3), ls='--', color = text_color)

# chart formating  
ax.set_xticks(ticks=[])
ax.set_yticks(ticks=[])
ax.spines[:].set_visible(False)


# to create name and value for each line of the main data 
for i, row in df_2000_mean_gas_price.iterrows(): 
    ax.text(
        x = row['avg_gas_price'] - 0.25, 
        y = i + 0.2,
        s = f"{row['avg_gas_price']:.1f}",
        color = text_color,
        ha = 'left',
        size = 12 
    )

    ax.text(
        x = 0, 
        y = i+ 0.2,  
        s = row['index'], 
        color = text_color, 
        size = 12 
    )

plt.tight_layout()
plt.show()
```

```python
df2['month_year'] = df2['Date'].dt.strftime('%Y-%m')
```

```python
df2
```

```python
aggregated_data = df2.groupby(['month_year']).agg(
    avg_price=('U.S.','mean')
)

aggregated_data.reset_index(inplace=True)

aggregated_data
```

```python
aggregated_data['month_year']
```

```python
fig, ax = plt.subplots()
ax.plot(aggregated_data['month_year'],aggregated_data['avg_price'], data=aggregated_data, linestyle = '-')
ax.tick_params(length=0)


location = ['1990-08','2000-08']
labels = ["$20k", 'T']
ax.set_xticks(location, labels=labels)


# ax.set_xticks([])
plt.show()

# note down approach to make this chart look cleaner  
# search for the period where the gas price spike up => check for the reasons that impacting on the gas price? => is there anything from public data that can lead to the volatile of this data. For example: oil production  
# search for chart ideas then make them as a learning approach 
```
